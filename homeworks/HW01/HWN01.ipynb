Примеры создания массивов

```python
#1
import numpy as np

#2
a = np.array([1, 2, 3])      # одномерный массив из списка
print ("a:Одномерный массив из списка:", a) 

b = np.array([[1.5, 2, 3], [4, 5, 6]])   # двумерный массив из списков
print ("b:Двумерный массив из списка:\n", b) 

aa=np.zeros((6, 9)) # двумерный массив из нулей, размером 6 на 9
print ("aa:Двумерный массив из нулей с использованием .zeros:\n", aa) 

bb=np.ones((4, 6)) # двумерный массив из нулей, размером 4 на 6
print ("bb:Двумерный массив из нулей с использованием .zeros:\n", bb)

cc=np.arange(1, 35, 2) # одномерный массив - последовательность от 1 до 35 с шагом 2
print ("cc:Одномерный массив из последовательности с использованием arange:\n", cc)

#3
print("bb:Форма двумерного массива:", bb.shape)      # shape получает форму двумерного массива
print("bb:Тип данных двумерного массива:", bb.dtype)    # dtype получает тип данных в заданном массиве

#4
print("cc:Первый элемент:", cc[0])             # обращение по индексу
print("cc:Последний элемент:", cc[-1])        # обращение по индексу с конца

print("cc:Элементы с 4 по 7-1 индекс (с нуля):", cc[4:7])    # срезы
print("cc:Элементы в обратном порядке:", cc[::-1]) 

#5
print("a + b =\n", a + b)          # сумма одномерного массива с двумерным

print("cc:Среднее значение:", cc.mean())    # среднее значение массива
print("cc:Минимальное значение:", cc.min())       # минимальный элемент
print("cc:Максимальное значение:", cc.max())        # максимальный элемент
print("cc:Сумма всех элементов:", cc.sum())        # сумма элементов
```
